## 🧩 Importing Core Libraries  

This step imports the main libraries required for fine-tuning and monitoring the model.  
Each library plays a key role in handling model operations, dataset management, and experiment tracking.

```python
from unsloth import FastLanguageModel
import torch
from datasets import load_dataset
from huggingface_hub import login
import wandb
```

## 🔐 Authentication & Token Setup  

This step securely connects Google Colab to your **Hugging Face** and **Weights & Biases (W&B)** accounts.  
It retrieves stored API tokens from Colab’s secret storage, logs in to both platforms, and initializes a new W&B tracking session for monitoring training progress.


```python
from google.colab import userdata
from huggingface_hub import login
import wandb

# Fetch tokens from Colab secret storage
hf_token = userdata.get("HGFaceApi")
wb_token = userdata.get("wb_token")

# Hugging Face login (required)
if not hf_token:
    raise ValueError("Hugging Face token not found in Colab secrets (key: 'HGFaceApi'). Please set it via userdata.set().")
login(hf_token)

# Weights & Biases login (optional: fall back to anonymous)
if wb_token:
    wandb.login(key=wb_token, relogin=True)
else:
    print("⚠️ W&B token ('wb_token') not found. Proceeding with anonymous logging.")

# Initialize a W&B run
run = wandb.init(
    project="LoRA-Finetuning-SmolLM2-135M",
    job_type="training",
    anonymous="allow"  # ensures run creation even without wb_token
)
```

## 🧠 Loading the Base Model and Tokenizer  

This step loads the **SmolLM2-135M** model and its tokenizer using Unsloth’s optimized `FastLanguageModel` class.  
The model is loaded in **4-bit precision** to reduce GPU memory usage, which is ideal for LoRA-based fine-tuning on Colab.


```python
max_seq_length = 2048
dtype = None
load_in_4bit = True  # LoRA supports 4-bit quantization to save VRAM

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/smollm2-135m",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    token = hf_token,
)
```

## 🧩 Adding LoRA Adapters  

This step attaches **LoRA (Low-Rank Adaptation)** layers to specific projection modules within the model.  
LoRA enables efficient fine-tuning by updating only a small set of parameters, significantly reducing memory usage while preserving model performance.

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,  # LoRA rank
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0.05,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 42,
    use_rslora = False,
    loftq_config = None,
)

print("✅ LoRA adapters added successfully!")
```

## 📚 Preparing and Formatting the Dataset  

This step loads a small subset of the **Alpaca dataset** and reformats each example into a structured **instruction–input–response** prompt.  
The formatting ensures the model learns to generate well-organized, context-aware responses while maintaining consistency across all training samples.

```python
dataset = load_dataset("tatsu-lab/alpaca", split="train[:500]")

prompt_style = """Below is an instruction that describes a task, paired with an input that provides further context.
Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token

def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs = examples["input"]
    outputs = examples["output"]
    texts = []
    for inst, inp, out in zip(instructions, inputs, outputs):
        if inp.strip() == "":
            text = prompt_style.format(inst, "N/A", out) + EOS_TOKEN
        else:
            text = prompt_style.format(inst, inp, out) + EOS_TOKEN
        texts.append(text)
    return {"text": texts}

dataset = dataset.map(formatting_prompts_func, batched=True)
print("✅ Sample formatted data:\n", dataset["text"][0][:400])
```

## 🏋️ Configuring the Trainer for Fine-Tuning  

This step initializes the **SFTTrainer** from the TRL library to manage the supervised fine-tuning process.  
It defines key training parameters such as batch size, learning rate, number of epochs, and precision mode (FP16 or BF16) to ensure efficient LoRA-based optimization and seamless integration with Weights & Biases for live progress tracking.

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    args = TrainingArguments(
        per_device_train_batch_size = 8,
        gradient_accumulation_steps = 1,
        num_train_epochs = 3,
        warmup_steps = 5,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 5,
        output_dir = "outputs",
        save_total_limit = 1,
        report_to = "wandb"
    ),
)
```

## 🚀 Starting the Fine-Tuning Process  

This step verifies the available GPU resources and begins the LoRA fine-tuning process.  
During training, performance metrics such as loss, runtime, and GPU utilization are automatically logged, and progress is tracked in real time through Weights & Biases.  
Once complete, the model is fully fine-tuned and ready for evaluation or deployment.

```python
# Display the GPU model and available VRAM
gpu_info = torch.cuda.get_device_properties(0)
print(f"🖥️ GPU: {gpu_info.name} | Total Memory: {round(gpu_info.total_memory / 1e9, 2)} GB")

# Begin training and store performance metrics
trainer_stats = trainer.train()

print("✅ LoRA fine-tuning completed successfully!")
```

## 📊 Training Summary & Resource Utilization  

This step reports key performance metrics after fine-tuning completes.  
It displays the total training runtime (in minutes) and the peak GPU memory reserved during the process, helping evaluate efficiency and hardware utilization for future training runs.

```python
used_mem = round(torch.cuda.max_memory_reserved() / 1e9, 3)
print(f"⏱ Runtime: {round(trainer_stats.metrics['train_runtime']/60, 2)} minutes")
print(f"💾 Peak reserved GPU memory: {used_mem} GB")
```

## 🔮 Inference & Model Evaluation  

This step switches the fine-tuned model to **inference mode** and tests its ability to generate responses.  
A sample Python-related instruction is provided to evaluate how well the model follows prompts and produces coherent, task-specific output after fine-tuning.

```python
from IPython.display import Markdown

FastLanguageModel.for_inference(model)

prompt = prompt_style.format(
    "Write a Python function to check if a number is even.",
    "",
    ""
)

inputs = tokenizer([prompt], return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_new_tokens=150)
Markdown(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

## ☁️ Uploading the Fine-Tuned Model to Hugging Face Hub  

This step saves the LoRA fine-tuned model and tokenizer locally, then uploads them to your Hugging Face account.  
It verifies your authentication, automatically adjusts the repository name to match your username, and creates a new model repository if one doesn’t already exist.  
Once completed, the fine-tuned model becomes accessible directly through the Hugging Face Hub for inference, sharing, or further experimentation.

```python

from huggingface_hub import HfApi

api = HfApi()

local_dir = "SmolLM2-135M-Alpaca-LoRA"
desired_repo_id = "Yugm1312/SmolLM2-135M-Alpaca-LoRA"  # your target path

# Save locally
model.save_pretrained(local_dir)
tokenizer.save_pretrained(local_dir)

# Resolve the authenticated username
who = api.whoami(token=hf_token)
username = who["name"]

# If token can't write to the desired namespace, switch to your own
namespace, repo_name = desired_repo_id.split("/", 1)
if namespace != username:
    print(f"ℹ️ Using token for '{username}', not '{namespace}'. "
          f"Pushing to '{username}/{repo_name}' instead.")
    repo_id = f"{username}/{repo_name}"
else:
    repo_id = desired_repo_id

# Create (or reuse) repo, then push
api.create_repo(repo_id=repo_id, repo_type="model", private=False, exist_ok=True, token=hf_token)
model.push_to_hub(repo_id, token=hf_token)
tokenizer.push_to_hub(repo_id, token=hf_token)

print(f"✅ LoRA fine-tuned model uploaded to https://huggingface.co/{repo_id}")
```
